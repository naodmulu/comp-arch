```python
# program that implements microprogrammed control,vertically
# user provieds the sepecification of the microprogram: the number of registers,number of ALU operations,an instruction to execute
# Display the final control word (rows of bits representing the respective microoperations)
# Display the final state of the registers

# import the necessary packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
# function to recieve the input from the user and return the number of registers and ALU operations to be performed
def get_input():
    # get the number of registers
    num_reg = int(input('Enter the number of registers: '))
    # get the number of ALU operations
    num_alu = int(input('Enter the number of ALU operations: '))
    # get the instruction to be executed
    instruction = input('Enter the instruction to be executed: ')
    return num_reg, num_alu, instruction
# function to create the microprogram
def create_microprogram(num_reg, num_alu):
    # create an empty list to store the microprogram
    microprogram = []
    # loop over the number of registers
    for i in range(num_reg):
        # get the microoperation for the register
        microop = input('Enter the microoperation for register {}: '.format(i+1))
        # append the microoperation to the microprogram
        microprogram.append(microop)
    # loop over the number of ALU operations
    for i in range(num_alu):
        # get the microoperation for the ALU
        microop = input('Enter the microoperation for ALU {}: '.format(i+1))
        # append the microoperation to the microprogram
        microprogram.append(microop)
    return microprogram
# function to create the control word
def create_control_word(microprogram, instruction):
    # create an empty list to store the control word
    control_word = []
    # loop over the microprogram
    for microop in microprogram:
        # check if the microoperation is in the instruction
        if microop in instruction:
            # append 1 to the control word
            control_word.append(1)
        else:
            # append 0 to the control word
            control_word.append(0)
    return control_word
# function to display the control word rows of bits representing the respective microoperations
def display_control_word(control_word):
    # loop over the control word
    for i in range(len(control_word)):
        # print the control word
        print('Control word {}: {}'.format(i+1, control_word[i]))
# function that implements microprogrammed control,vertically
def microprogrammed_control(num_reg, num_alu, instruction):
    # create the microprogram
    microprogram = create_microprogram(num_reg, num_alu)
    # create the control word
    control_word = create_control_word(microprogram, instruction)
    # display the control word
    display_control_word(control_word) 




```
